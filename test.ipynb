```python
import pandas as pd
import numpy as np
import glob
from math import *

#Takes path as input and gives time column in csv file as output list
def npArray(lst):
    temp = np.array(lst)
    return temp

def dataExtractA(lst):
    temp = pd.read_csv(lst)
    temp2 = []
    temp3 = []
    for i in range(temp.shape[0]):
        temp2.append(temp.time[i])
        temp3.append(temp.vol[i])
    return np.array(temp2),np.array(temp3)

def dataExtractB(lst):
    temp = pd.read_csv(lst)
    temp2 = []
    temp3 = []
    temp4 = []
    temp5 = []
    temp6 = []
    temp7 = []
    temp8 = []
    for i in range(temp.shape[0]):
        temp2.append(temp.time[i])
        temp3.append(temp.vol[i])
        temp4.append(temp.ph[i])
        temp5.append(temp.solids[i])
        temp6.append(temp.hardness[i])
        temp7.append(temp.oil[i])
        temp8.append(temp.bod[i])
    return np.array(temp2),np.array(temp3),np.array(temp4),np.array(temp5),np.array(temp6),np.array(temp7),np.array(temp8)

def logic(data1,data2,comp):
    #Paths for A and B
    #data1 = glob.glob('1/HCP4A.csv')
    #data2 = glob.glob('1/HCP4B.csv')
    
    #Initialize data list
    dataATime = []
    dataAVol = []
    
    dataBTime = []
    dataBVol = []
    dataBPh = []
    dataBSolids = []
    dataBHardness = []
    dataBOil = []
    dataBBod = []
    
    #Converts to numpy array and store in dataA and dataB
    dataATime,dataAVol = dataExtractA(data1)
    dataBTime,dataBVol,dataBPh,dataBSolids,dataBHardness,dataBOil,dataBBod = dataExtractB(data2)
    
    #Initialize new clean data list
    dataATimeC = []
    dataAVolC = []
    
    dataBTimeC = []
    dataBVolC = []
    dataBPhC = []
    dataBSolidsC = []
    dataBHardnessC = []
    dataBOilC = []
    dataBBodC = []
    
    #Take min of size of dataA and dataB as size for iteration
    size = min(dataATime.shape[0],dataBTime.shape[0])
    #print(size,dataATime.shape[0],dataBTime.shape[0])
    #Iterate "size" times
    at=0
    bt=0
    
    for i in range(size):
        #print(at+i>size)
        #nb+=NB
        #Difference betweent time in dataA and time in DataB should be less then 900000.0 (15 minutes)
        if(abs(dataATime[i-at]-dataBTime[i-bt])<900000.0):
            #Append these values to clean data lists for A and B
            dataATimeC.append(dataATime[i-at])
            dataAVolC.append(dataAVol[i-at])
    
            dataBTimeC.append(dataBTime[i-bt])
            dataBVolC.append(dataBVol[i-bt])
            dataBPhC.append(dataBPh[i-bt])
            dataBSolidsC.append(dataBSolids[i-bt])
            dataBHardnessC.append(dataBHardness[i-bt])
            dataBOilC.append(dataBOil[i-bt])
            dataBBodC.append(dataBBod[i-bt])
        elif(dataATime[i-at]-dataBTime[i-bt]<-900000.0):
            bt=bt+1
        else:
            print("A is missing for B value")
    
    #Testing values of cleanA and cleanB data
    size = min(np.asarray(dataATimeC).shape[0],np.asarray(dataBTimeC).shape[0])
    for i in range(size):
        if(abs(dataATimeC[i]-dataBTimeC[i])>900000.0):
            #Should not print this line
            print("Algorithm is failing")
    
    dataATimeC = npArray(dataATimeC)
    dataAVolC = npArray(dataAVolC)
    
    dataBTimeC = npArray(dataBTimeC)
    dataBVolC = npArray(dataBVolC)
    dataBPhC = npArray(dataBPhC)
    dataBSolidsC = npArray(dataBSolidsC)
    dataBHardnessC = npArray(dataBHardnessC)
    dataBOilC = npArray(dataBOilC)
    dataBBodC = npArray(dataBBodC) 
    
    #Model 1
    pctg=[]
    total=0
    counter=0
    for i in range(np.asarray(dataATimeC).shape[0]):
        if((dataBVolC[i]!=0.0) & (dataAVolC[i]!=0.0)):
            pcg=dataBVolC[i]/dataAVolC[i]
            pctg.append(pcg)
            total=total+pcg
            counter=counter+1
    if(total!=0.0):
        avgpctg=total/counter
    else:
        avgpctg=0.00
    counter=0
    for i in range(np.asarray(dataATimeC).shape[0]):
        if((dataBVolC[i]!=0.0) & (dataAVolC[i]!=0.0)):
            if(abs(avgpctg-pcg)>0.15):
                #print(avgpctg,pcg)
                counter=counter+1

    #Model 2
    pctg=[]
    total=0
    counter1=0
    for i in range(np.asarray(dataATimeC).shape[0]):
        if((dataBVolC[i]!=0.0) & (dataAVolC[i]!=0.0)):
            pcg=dataBVolC[i]/dataAVolC[i]
            pctg.append(pcg)
            total=total+pcg
            counter1=counter1+1
    if(total!=0.0):
        avgpctg=total/counter1
    else:
        avgpctg=0.00
    flag1=False
    if(avgpctg>0.45):
        flag1=True
        
    """
    #Model 3
    pctg=[]
    total=0
    counter2=0
    for i in range(np.asarray(dataATimeC).shape[0]):
        if((dataBVolC[i]!=0.0) & ((dataBPhC[i]!=0.0) | (dataBSolidsC[i]!=0.0) | (dataBHardnessC[i]!=0.0) | (dataBSolidsC[i]!=0.0) | (dataBOilC[i]!=0.0) | (dataBBodC[i]!=0.0) )):
            flag2=True
        else:
            flag2=False
    """
    print("{}      {}       {}".format(comp,counter,flag1))
    

#NBb=+n
print("Name   Frequeny    Threshold_Crossed")
data1A = glob.glob('1/????A.csv')
data2A = glob.glob('1/?????A.csv')
data3A = glob.glob('1/??????A.csv')
data1B = glob.glob('1/????B.csv')
data2B = glob.glob('1/?????B.csv')
data3B = glob.glob('1/??????B.csv')
counter=0
for i in range(len(data1A)):
    counter=counter+1
    logic(data1A[i],data1B[i],counter)
for i in range(len(data2A)):
    counter=counter+1
    logic(data2A[i],data2B[i],counter)
for i in range(len(data3A)):
    counter=counter+1
    logic(data3A[i],data3B[i],counter)
```
